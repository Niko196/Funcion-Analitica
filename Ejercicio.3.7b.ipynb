## Otra forma de ver Y(x)

```python
def cifras(x):
    x *= 100
    x = int(x)
    x /= 100
    return x

x = 1.37
y = ((x-7)*x+8)*x+0.35
yapr = cifras((cifras(cifras(x-7)*x)+8)*x)+0.35
et = (y-yapr)
er = et/yapr*100
es = (0.5*10**(2-3))
print ("Y\t\t|\tY aprox.\t|\tEt\t\t|\tE% aprox.\t|\tEs")
print ("-------------------------------------------------------------"
      +"------------------------------------------------")
print ("%f\t|\t%f\t|\t%f\t|\t%f\t|\t%f" %(y, yapr, et, er, es))
```
