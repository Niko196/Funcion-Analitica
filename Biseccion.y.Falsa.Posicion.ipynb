```python
def funcion(x) :
    return ((x**10)-1)
    
xl, xu, xr, rep = 0.0, 1.3, 0.0, -1
print ("Itera.\t|\tXu\t\t|\tXr\t\t|\tXl\t\t|\tF(x)")
print ("----------------------------------------------------------"
      +"----------------------------------------------------------")
while True:
    rep += 1
    if (xr!=0):
        print ("%d\t|\t%f\t|\t%f\t|\t%f\t|\t%f" %(rep, xu, xr, xl, funcion(xr)))
    else :
        print ("%d\t|\t%f\t|\t%f\t|\t%f\t|\t-.---" %(rep, xu, xr, xl))
    if (funcion(xl)*funcion(xu)<0):
        xr = (xl+xu)/2
        if (funcion(xl)*funcion(xr)<0) :
            xu = xr
            continue
        if (funcion(xl)*funcion(xr)>0):
            xl = xr
            continue
        if (funcion(xl)*funcion(xr)==0):
            print ("La raiz esta en t = %1.3f" %(xr))
            break
    else :
        break

print ()

xl, xu, xr, rep =  0.0, 1.3, 0.0, -1
print ("Itera.\t|\tXu\t\t|\tXr\t\t|\tXl\t\t|\tF(x)")
print ("----------------------------------------------------------"
      +"----------------------------------------------------------")
while True:   ## Se queda en repeticion infinita debido a que nunca se cumple la condicion indicada
    rep += 1
    if (xr!=0):
        print ("%d\t|\t%f\t|\t%f\t|\t%f\t|\t%f" %(rep, xu, xr, xl, funcion(xr)))
    else :
        print ("%d\t|\t%f\t|\t%f\t|\t%f\t|\t-.---" %(rep, xu, xr, xl))
    if (funcion(xl)*funcion(xu)<0):
        xr = (xu-((funcion(xu)*(xl-xu))/(funcion(xl)-funcion(xu))))
        if (funcion(xl)*funcion(xr)<0):  ## Devido a que el resultado es muy pequeo y queda en ceros con el signo
            xu = xr
            continue
        if (funcion(xl)*funcion(xr)>0):  ## del numero, en este caso el menos, el cual afecta en la decicion del criterio
            xl = xr
            continue
        if (funcion(xl)*funcion(xr)==0):  ## <<---
            print ("La raiz esta en t = %1.3f" %(xr))
            break
    else:
        break
```
